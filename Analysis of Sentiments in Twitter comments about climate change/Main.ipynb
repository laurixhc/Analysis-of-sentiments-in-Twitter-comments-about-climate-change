This are the different things we need to use from the different libraries.

```python
# To manage and manipulate data
import pandas as pd
import numpy as np

# For text preprocessing
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# For sentiment analysis
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# For data visualization
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

from collections import Counter
from IPython.display import display

# Download necessary nltk resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Upload CSV file
arch = pd.read_csv('ClimateChange.csv')
```

Function to clean the tweets, deleting the special characters, and converting the words to lowercase.

```python
import re

def clean_tweet(tweet):
    """
    Clean up the tweet text by removing links, mentions, hashtags and special characters.
    
    Parameters:
    tweet (str): tweet text.
    
    Return:
    str: tweet text cleaned.
    """
    tweet = re.sub(r'http\S+', '', tweet)  # Delete URLs
    tweet = re.sub(r'@\w+', '', tweet)    # Delete mentions
    tweet = re.sub(r'#\w+', '', tweet)    # Delete hashtags
    tweet = re.sub(r'[^\w\s]', '', tweet) # Delete special characters
    tweet = tweet.lower()                 # Convert to lowercase
    return tweet

# Apply the cleanup function to the text column of tweets
arch['cleaned_tweet'] = arch['Embedded_text'].apply(clean_tweet)
```

## 6.1. Which one is the sentiment of each tweet?

Analysing sentiments will give us a feedback about how are people feeling. This will help us to know how go act, to know if people feel that there is any improvement or if not. In this type of topics, analysing people sentiments is essential, since we need to understand public opinion, for marketing, politic and social issues. Also, activism and movements will help to solve the problem, since more people will be involved, and the more people involved, the greater the results will be reflected to avoid climate change.

Here we have created a function which analize a given tweet sentiment, by using TextBlob. After it, we are going to give all the tweets of our file, and we are going to apply it, creating a new feature, 'sentiment'. Then we calculate which one is the percentage of each sentiment and we print it.

```python
def analyze_sentiment(tweet):
    """
    Analyze tweet sentiment using TextBlob.
    
    Parameters:
    tweet (str): tweet text.
    
    Return:
    str: Polarity of sentiment (positive, negative or neutral).
    """
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

# Apply sentiment analysis feature to clean tweets column
arch['sentiment'] = arch['cleaned_tweet'].apply(analyze_sentiment)

# Calculate the percentage of each type of feeling
counter = arch['sentiment'].value_counts()
total = len(arch)
porc = (counter / total) * 100
porc_dict = porc.to_dict()

# We print the porcentage of each sentiment
print(porc_dict)
```

We can see that the majority of tweets are positive, more than half, a quarter are negative, and the rest are neutral. Interpreting the result, we can observe that the Twitter society has positive thinking about what is happening with climate change, that is, they notice changes, they think that the changes can end up stopping climate change.

We can see that my hypothesis was wrong. This shows that many times what one thinks is not the right thing.

Now we are going to see this information represented, in a bars graphic.

```python
# Count the number of tweets per sentiment
sentiment_bars = arch['sentiment'].value_counts()

# To create a bars graphic
plt.figure(figsize=(12, 6))
sns.barplot(x=sentiment_bars.index, y=sentiment_bars.values)
plt.title('Sentiments distribution')
plt.xlabel('Sentiments')
plt.ylabel('Tweets quantity')
plt.show()
```

We can see that is really big the difference between the positive sentiment and the other two sentiments. Between the negative and the neutral, there is not that much difference.

## 6.3. Which ones are the most used words in negative sentiment tweets?

This will provide us with information about the most used topics when making a tweet whose sentiment is negative, because if a word is widely used and is also representative of an approach, it may mean the population's concern about that topic.

To represent this information, we will use a word cloud. The larger the word, the more times it appears, and vice versa.

```python
#we take the negative tweets cleaned
negative = arch[arch['sentiment'] == 'negative']['cleaned_tweet']

#we put all words in english
stopWords = set(stopwords.words('english'))
tokens = [word for tweet in negative for word in tweet.split() if word not in stopWords]

# we count the frequency of each word
freq = Counter(tokens)

#To create the cloud
cloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq)

plt.figure(figsize=(10, 5))
plt.imshow(cloud, interpolation='bilinear')
plt.title('Word Cloud for Negative Tweets')
plt.axis('off')
plt.show()
```

We can see that the most used words are climate change, replying, tweet, ... and also some numbers. But the numbers are not important information for us, since we can not get any conclusion by knowing that it is very used. This is why, to make the selection of the ten most used words, we eliminate the numbers, since a word will give us much more information than a number.

These are the ten most used words in this context.

```python
# We take all words except numbers
tokensNNumb = [word for word in tokens if not word.isdigit()]

# We count how much appears each word
freqNNumb = Counter(tokensNNumb)

tenWords = freqNNumb.most_common(10)
words, counts = zip(*tenWords)

# To create the data frame
common = pd.DataFrame(list(tenWords), columns=['word', 'count'])

plt.figure(figsize=(10, 5))
sns.barplot(data=common, x='count', y='word')
plt.title('10 Most Common Words in Negative Tweets')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.show()
```

We can see that the most common words used in negative tweets are the ones above. The word that I observe that can give us more information is world, since they may be talking about concern for the world, because they feel that the world has not improved as much as it should.

## 6.5. Which ones are the most used hashtags?

Hashtags play a crucial role in organizing and categorizing content on social media platforms. This is way, analysing the most used hashtags will help to extract information about the different topics, inside the climate change topic.

By leveraging insights from hashtag analysis, individuals and organizations can make more informed decisions and improve their media strategies.

```python
# Function to extract hashtags
def extract_hashtags(tweet):
    hashtags = re.findall(r'#\w+', tweet)
    return hashtags

arch['hashtags'] = arch['Embedded_text'].apply(extract_hashtags)

# Flatten the list of hashtag lists and count the frequency of each one
hashtags = [hashtag for hashtags_list in arch['hashtags'] for hashtag in hashtags_list]
freqHash = Counter(hashtags)

# Show the 10 most common hashtags
commonHash = freqHash.most_common(10)
tableHash = pd.DataFrame(freqHash.most_common(10), columns=['hashtag', 'count'])
display(tableHash.style.hide(axis='index').set_table_attributes('border="1" class="dataframe table table-hover table-bordered"').set_caption("Most frequent hashtags"))
```

We can see that most hashtags are related with climate change, crisis, ...

On the other hand, there are some hashtags, such as #auspol or #insiders, which are related with politic, in this case, this one means Australian Politics. This can be because there are lot of tweets about the australian politic elections.

## 6.6. Which ones are the tweets with more retweets and likes?

Knowing which tweets have the most retweets and the most likes is essential to better understand the dynamics of popularity and interaction on Twitter. Provides a clear view of audience preferences and behaviors.

Analyzing this data allows individuals and organizations to make decisions and continually improve their social media presence.

```python
# Function to clean the text
def clean(text):
    pattern = r'\s+\d+\s+\d+\s+\d+$'  # This expression looks for three groups of numbers at the end of the text
    return re.sub(pattern, '', text)

# Function to convert data to numerical values, wihtout letters and ','
def convert_data(value):
    if pd.isna(value):
        return(np.nan)
    value = str(value)
    if all(char not in value for char in [',', 'K', 'M']):
        return value
    else:
        if ',' in value:
            value = value.replace(',', '')
        if 'K' in value:
            value = float(value.replace('K', ''))* 1000
        elif 'M' in value:
            value = float(value.replace('M', '')) * 1000000
        try:
            return float(value)
        except ValueError:
            return np.nan

# Sort the DataFrame by the retweets column in descending order

arch['Embedded_text'] = arch['Embedded_text'].replace('\n', ' ', regex=True).replace('\r', ' ', regex=True).replace('\t', ' ', regex=True).str.strip().apply(clean)

arch['Retweets'] = arch['Retweets'].apply(convert_data)

arch['Retweets'] = pd.to_numeric(arch['Retweets'], errors='coerce')

archNoNaN = arch.dropna(subset=['Retweets'])

sortRetw = archNoNaN.sort_values(by='Retweets', ascending=False)

# Select the 10 tweets with the most retweets
tenSortRetw = sortRetw.head(10)
pd.set_option('display.max_colwidth', None)

# Show tweets with the most retweets
display(tenSortRetw[['Embedded_text', 'Retweets','sentiment']])

# We are going to make the same, but with likes instead of retweets.
arch['Likes'] = arch['Likes'].apply(convert_data)
arch['Likes'] = pd.to_numeric(arch['Likes'], errors='coerce')
archNoNaNLikes = arch.dropna(subset=['Retweets'])
sortLikes = archNoNaNLikes.sort_values(by='Likes', ascending=False)
tenSortLikes = sortLikes.head(10)

pd.set_option('display.max_colwidth', None)

# Show tweets with the most retweets
display(tenSortLikes[['Embedded_text', 'Likes','sentiment']])
```

We can see that the tweets with the most likes do not coincide with the ones with the most likes. It is true that there are several that are on both lists. Despite this, we see that very different topics are discussed, some more related to politics and others more related to the changes that have been reflected.

In the case of retweets, there are more negative ones. On the other hand, in the case of likes, there are more positive ones. It is a curious difference and worth analyzing.
